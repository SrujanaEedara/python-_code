```python
#OPERTIONS ON LINKED LISTS
class Node:
    def __init__(self,value):
        self.data=value
        self.next=None

head=Node(1)
head.next=Node(2)
head.next.next=Node(3)
head.next.next.next=Node(4)
head.next.next.next.next=Node(5)

print(head.data,"-->",
head.next.data,"-->",
head.next.next.data,"-->",
head.next.next.next.data,"-->",
head.next.next.next.next.data,"-->",)
#start insert
x=int(input())
temp=Node(x)
temp.next=head
head=temp
print(head.data)

#to display
print(head.data,"-->",
head.next.data,"-->",
head.next.next.data,"-->",
head.next.next.next.data,"-->",
head.next.next.next.next.data,"-->",
head.next.next.next.next.next.data,"-->")

#end insert
x=int(input())
temp=Node(x)
curr=head
if curr==None:
    head=temp
else:
    while curr.next!=None:
        curr=curr.next
    curr.next=temp 
    
#to display
curr=head
while curr!=None:
    print(curr.data,"-->",end=" ")
    curr=curr.next

```

```python
#TREES
class node:
    def __init__(self,value):
        self.data=value
        self.left=None
        self.right=None
#PREORDER TRAVERSAL
def preorder(root):
    if root==None:
        return
    print(root.data)
    preorder(root.left)
    preorder(root.right)

#INORDER TRAVERSAL
def inorder(root):
    if root==None:
        return
    inorder(root.left)
    print(root.data)
    inorder(root.right)
    
#POSTORDER TRAVERSAL
def postorder(root):
    if root==None:
        return
    postorder(root.left)
    postorder(root.right)
    print(root.data)

#LEVELORDER TRAVERSAL
def levelorder(root):
    Q=[root,None]
    while Q:
        curr=Q.pop(0)
        if curr==None:
            print()
            if Q:
               Q.append(curr)
        else:
            print(curr.data,end=" ")
            if curr.left!=None:
                Q.append(curr.left)
            if curr.right!=None:
                Q.append(curr.right)
    

root=node(1)
root.left=node(2)
root.right=node(3)

root.left.left=node(4)
root.left.right=node(5)

root.right.left=node(6)
root.right.right=node(7)

print(" elements in preorder traversal :\n")
preorder(root)
print("elements in inorder traversal :\n")
inorder(root)
print("elements in postorder traversal :\n")
postorder(root)
print("elements in levelorder traversal :\n")
levelorder(root)
```
